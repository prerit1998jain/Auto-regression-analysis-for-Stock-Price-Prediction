```python
import matplotlib.pyplot as plt
plt.figure(num=None, figsize=(20, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(data.index[10000:],data[10000:].High.rolling(15).sum(),data.index[10000:],data[10000:].Low.rolling(15).sum())

plt.title('Data')
plt.grid(True)
plt.show()
#ye_2017_normal = year_2017[:-35]
```

```python
data['range'] = data.High - data.Low
```

```python
import matplotlib.pyplot as plt
plt.figure(num=None, figsize=(20, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(data[11650:].range.rolling(15).sum())
plt.title('Data')
plt.grid(True)
plt.show()
#ye_2017_normal = year_2017[:-35]
```

```python
copy = list(data[:12073].High.values)
difference = list(data.High[1:] - copy)
difference.append(0)
data['difference'] = difference
```

```python
import matplotlib.pyplot as plt
plt.figure(num=None, figsize=(20, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(data[10000:].difference.rolling(15).sum())
plt.title('Data')
plt.grid(True)
plt.show()
```

```python
from pandas import read_csv
from statsmodels.tsa.stattools import adfuller
X = data.High.values
result = adfuller(X[:])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
	print('\t%s: %.3f' % (key, value))
```
